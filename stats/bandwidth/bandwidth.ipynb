```python
# Import the modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Read all the files
datasets = {}
datasets["1"] = pd.DataFrame(pd.read_csv("bw_1.csv"))
datasets["2"] = pd.DataFrame(pd.read_csv("bw_2.csv"))
datasets["4"] = pd.DataFrame(pd.read_csv("bw_4.csv"))
datasets["8"] = pd.DataFrame(pd.read_csv("bw_8.csv"))
datasets["16"] = pd.DataFrame(pd.read_csv("bw_16.csv"))
datasets["32"] = pd.DataFrame(pd.read_csv("bw_32.csv"))
datasets["64"] = pd.DataFrame(pd.read_csv("bw_64.csv"))
datasets["128"] = pd.DataFrame(pd.read_csv("bw_128.csv"))


X = list(datasets["1"].iloc[:, 0])
```

```python
# Plot number of invocations
Y = list(datasets["1"].iloc[:, 1])
plt.figure(figsize=(10,6))
plt.plot(X, Y, color='g')
plt.ylim(0)
plt.title("Invocations")
plt.xlabel("Second")
plt.ylabel("Number of invocations")

plt.show()
```

```python
# Plot disk memory
Y0 = list(datasets["1"].iloc[:, 8])
Y1 = list(datasets["2"].iloc[:, 8])
Y2 = list(datasets["4"].iloc[:, 8])
Y3 = list(datasets["8"].iloc[:, 8])
Y4 = list(datasets["16"].iloc[:, 8])
Y5 = list(datasets["32"].iloc[:, 8])
Y6 = list(datasets["64"].iloc[:, 8])
Y7 = list(datasets["128"].iloc[:, 8])

plt.plot(X, Y0, label="1 GB/s RW")
plt.plot(X, Y1, label="2 GB/s RW")
plt.plot(X, Y2, label="4 GB/s RW")
plt.plot(X, Y3, label="8 GB/s RW")
plt.plot(X, Y4, label="16 GB/s RW")
plt.plot(X, Y5, label="32 GB/s RW")
plt.plot(X, Y6, label="64 GB/s RW")
plt.plot(X, Y7, label="128 GB/s RW")
plt.title("Disk storage cache (250 GB)")
plt.xlabel("Second")
plt.ylabel("Disk storage being used (MB)")
plt.legend()
plt.show()
```

```python
# Plot cold starts
Y0 = list(datasets["1"].iloc[:, 3])
Y1 = list(datasets["2"].iloc[:, 3])
Y2 = list(datasets["4"].iloc[:, 3])
Y3 = list(datasets["8"].iloc[:, 3])
Y4 = list(datasets["16"].iloc[:, 3])
Y5 = list(datasets["32"].iloc[:, 3])
Y6 = list(datasets["64"].iloc[:, 3])
Y7 = list(datasets["128"].iloc[:, 3])

plt.plot(X, Y0, label="1 GB/s R/W")
plt.plot(X, Y1, label="2 GB/s R/W")
plt.plot(X, Y2, label="4 GB/s R/W")
plt.plot(X, Y3, label="8 GB/s R/W")
plt.plot(X, Y4, label="16 GB/s R/W")
plt.plot(X, Y5, label="32 GB/s R/W")
plt.plot(X, Y6, label="64 GB/s R/W")
plt.plot(X, Y7, label="128 GB/s R/W")
plt.ylim(0)
plt.title("Cold Starts (250GB Disk | 10GB RAM)")
plt.xlabel("Second")
plt.ylabel("Number of cold invocations")
plt.legend()
plt.show()
```

```python
#Plot lukewarm starts vs disk
data = pd.read_csv("lukewarm.csv")
df = pd.DataFrame(data)
X = list(df.iloc[:, 0])
Y = list(df.iloc[:, 1])
plt.scatter(X, Y, color="red")
plt.ylim(0)
plt.title("Luke warm starts vs average bandwidth (250GB Disk cache | 10GB RAM cache)")
plt.xlabel("Average bandwidth (MB/s)")
plt.ylabel("Percentage of lukewarm starts")
```

```python
#Plot latency CDFs
data_1 = pd.read_csv("latencies_bw_1.csv")
data_2 = pd.read_csv("latencies_bw_2.csv")
data_4 = pd.read_csv("latencies_bw_4.csv")
data_8 = pd.read_csv("latencies_bw_8.csv")
data_16 = pd.read_csv("latencies_bw_16.csv")
data_32 = pd.read_csv("latencies_bw_32.csv")
data_64 = pd.read_csv("latencies_bw_64.csv")
data_128 = pd.read_csv("latencies_bw_128.csv")
df_1 = pd.DataFrame(data_1)
df_2 = pd.DataFrame(data_2)
df_4 = pd.DataFrame(data_4)
df_8 = pd.DataFrame(data_8)
df_16 = pd.DataFrame(data_16)
df_32 = pd.DataFrame(data_32)
df_64 = pd.DataFrame(data_64)
df_128 = pd.DataFrame(data_128)
x1 = np.sort(list(df_1.iloc[:, 0]))
y1 = np.arange(len(df_1)) / float(len(df_1))
x2 = np.sort(list(df_2.iloc[:, 0]))
y2 = np.arange(len(df_2)) / float(len(df_2))
x4 = np.sort(list(df_4.iloc[:, 0]))
y4 = np.arange(len(df_4)) / float(len(df_4))
x8 = np.sort(list(df_8.iloc[:, 0]))
y8 = np.arange(len(df_8)) / float(len(df_8))
x16 = np.sort(list(df_16.iloc[:, 0]))
y16 = np.arange(len(df_16)) / float(len(df_16))
x32 = np.sort(list(df_32.iloc[:, 0]))
y32 = np.arange(len(df_32)) / float(len(df_32))
x64 = np.sort(list(df_64.iloc[:, 0]))
y64 = np.arange(len(df_64)) / float(len(df_64))
x128 = np.sort(list(df_128.iloc[:, 0]))
y128 = np.arange(len(df_128)) / float(len(df_128))

# plotting
plt.xlabel('latency (ms)')
plt.ylabel('percentile')
  
plt.title('Latency CDF')
  
plt.plot(x1, y1, label="1 R/W GB/s")
plt.plot(x2, y2, label="2 R/W GB/s")
plt.plot(x4, y4, label="4 R/W GB/s")
plt.plot(x8, y8, label="8 R/W GB/s")
plt.plot(x16, y16, label="16 R/W GB/s")
plt.plot(x32, y32, label="32 R/W GB/s")
plt.plot(x64, y64, label="64 R/W GB/s")
plt.plot(x128, y128, label="128 R/W GB/s")
plt.ylim(0)
plt.xlim([0,400])
plt.legend()
plt.show()
```
