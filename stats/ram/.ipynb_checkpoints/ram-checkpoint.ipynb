```python
# Import the modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Read all the files
datasets = {}
datasets["10_250"] = pd.DataFrame(pd.read_csv("10_250_3000_ram.csv"))
datasets["5_250_ram"] = pd.DataFrame(pd.read_csv("5_250_3000_ram.csv"))
datasets["15_250_ram"] = pd.DataFrame(pd.read_csv("5_250_3000_ram.csv"))
datasets["20_250_ram"] = pd.DataFrame(pd.read_csv("20_250_3000_ram.csv"))
datasets["10_50_disk"] = pd.DataFrame(pd.read_csv("10_50_3000_disk.csv"))
datasets["10_100_disk"] = pd.DataFrame(pd.read_csv("10_100_3000_disk.csv"))
datasets["10_500_disk"] = pd.DataFrame(pd.read_csv("10_500_3000_disk.csv"))
datasets["10_1000_disk"] = pd.DataFrame(pd.read_csv("10_1000_3000_disk.csv"))


X = list(datasets["10_250"].iloc[:, 0])
```

```python
# Plot number of invocations
Y = list(datasets["10_250"].iloc[:, 1])
plt.figure(figsize=(10,6))
plt.plot(X, Y, color='g')
plt.title("Invocations")
plt.xlabel("Second")
plt.ylabel("Number of invocations")

plt.show()
```

```python
# Plot failed invocations
Y = list(datasets["10_250"].iloc[:, 2])

plt.bar(X, Y, color='r', width=1.0)
plt.title("Failed Invocations")
plt.xlabel("Second")
plt.ylabel("Number of failed invocations")

plt.show()
```

```python
# Plot cold starts
Y0 = list(datasets["10_250"].iloc[:, 3])
Y1 = list(datasets["5_250_ram"].iloc[:, 3])
Y2 = list(datasets["15_250_ram"].iloc[:, 3])
Y3 = list(datasets["20_250_ram"].iloc[:, 3])

plt.plot(X, Y0, color='b', label="10GB RAM cache")
plt.plot(X, Y1, color='r', label="5GB RAM cache")
plt.plot(X, Y2, color='g', label="15GB RAM cache")
plt.plot(X, Y3, color='orange', label="20GB RAM cache")
plt.title("Cold Starts")
plt.xlabel("Second")
plt.ylabel("Number of cold invocations")
plt.legend()
plt.show()
```

```python
# Plot warm starts
Y = list(datasets["10_250"].iloc[:, 4])

plt.bar(X, Y, color='y', width=1.0)
plt.title("Warm Starts")
plt.xlabel("Second")
plt.ylabel("Number of warm invocations")

plt.show()
```

```python
# Plot lukewarm starts
Y = list(datasets["10_250"].iloc[:, 7])

plt.bar(X, Y, color="orange", width=1.0)
plt.title("Lukewarm starts")
plt.xlabel("Second")
plt.ylabel("Number of lukewarm invocations")

plt.show()
```

```python
# Plot run memory
Y = list(datasets["10_250"].iloc[:, 5])

plt.bar(X, Y, color="pink", width=1.0)
plt.title("Run memory")
plt.xlabel("Second")
plt.ylabel("Run memory being used")

plt.show()
```

```python
# Plot ram memory
Y = list(datasets["10_250"].iloc[:, 6])

plt.bar(X, Y, color="purple", width=1.0)
plt.title("RAM cache memory")
plt.xlabel("Second")
plt.ylabel("Ram cache memory being used")

plt.show()
```

```python
#Plot warm starts vs RAM
data = pd.read_csv("warm.csv")
df = pd.DataFrame(data)
X = list(df.iloc[:, 0])
Y = list(df.iloc[:, 1])
plt.plot(X, Y, color="orange")
plt.ylim(0)
plt.title("RAM Cache size vs warm starts")
plt.xlabel("RAM cache size (GB)")
plt.ylabel("Number of warm starts")
```

```python
#Plot lukewarm starts vs disk
data = pd.read_csv("lukewarm.csv")
df = pd.DataFrame(data)
X = list(df.iloc[:, 0])
Y = list(df.iloc[:, 1])
plt.plot(X, Y, color="red")
plt.ylim(0)
plt.title("Disk cache size vs lukewarm starts")
plt.xlabel("Disk cache size (GB)")
plt.ylabel("Number of lukewarm starts")
```

```python
#Plot latency CDFs
data_50 = pd.read_csv("latencies_10_50_3000_disk.csv")
data_100 = pd.read_csv("latencies_10_100_3000_disk.csv")
data_250 = pd.read_csv("latencies_10_250_3000_disk.csv")
data_500 = pd.read_csv("latencies_10_500_3000_disk.csv")
data_1000 = pd.read_csv("latencies_10_1000_3000_disk.csv")
df_50 = pd.DataFrame(data_50)
df_100 = pd.DataFrame(data_100)
df_250 = pd.DataFrame(data_250)
df_500 = pd.DataFrame(data_500)
df_1000 = pd.DataFrame(data_1000)
x50 = np.sort(list(df_50.iloc[:, 0]))
y50 = np.arange(len(df_50)) / float(len(df_50))
x100 = np.sort(list(df_100.iloc[:, 0]))
y100 = np.arange(len(df_100)) / float(len(df_100))
x250 = np.sort(list(df_250.iloc[:, 0]))
y250 = np.arange(len(df_250)) / float(len(df_250))
x500 = np.sort(list(df_500.iloc[:, 0]))
y500 = np.arange(len(df_500)) / float(len(df_500))
x1000 = np.sort(list(df_1000.iloc[:, 0]))
y1000 = np.arange(len(df_1000)) / float(len(df_1000))

# plotting
plt.xlabel('x-axis')
plt.ylabel('y-axis')
  
plt.title('Latency CDF')
  
plt.plot(x50, y50, label="50GB Disk Cache")
plt.plot(x100, y100, label="100GB Disk Cache")
plt.plot(x250, y250, label="250GB Disk Cache")
plt.plot(x500, y500, label="500GB Disk Cache")
plt.plot(x1000, y1000, label="1000GB Disk Cache")
plt.ylim(0)
plt.xlim(0)
plt.legend()
plt.show()
```
