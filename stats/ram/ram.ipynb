```python
# Import the modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Read all the files
datasets = {}
datasets["2"] = pd.DataFrame(pd.read_csv("ram_2.csv"))
datasets["4"] = pd.DataFrame(pd.read_csv("ram_4.csv"))
datasets["8"] = pd.DataFrame(pd.read_csv("ram_8.csv"))
datasets["16"] = pd.DataFrame(pd.read_csv("ram_16.csv"))
datasets["32"] = pd.DataFrame(pd.read_csv("ram_32.csv"))
datasets["64"] = pd.DataFrame(pd.read_csv("ram_64.csv"))
datasets["128"] = pd.DataFrame(pd.read_csv("ram_128.csv"))


X = list(datasets["2"].iloc[:, 0])
```

```python
# Plot number of invocations
Y = list(datasets["2"].iloc[:, 1])
plt.figure(figsize=(10,6))
plt.plot(X, Y, color='g')
plt.ylim(0)
plt.title("Invocations")
plt.xlabel("Second")
plt.ylabel("Number of invocations")

plt.show()
```

```python
# Plot cold starts
Y0 = list(datasets["2"].iloc[:, 3])
Y1 = list(datasets["4"].iloc[:, 3])
Y2 = list(datasets["8"].iloc[:, 3])
Y3 = list(datasets["16"].iloc[:, 3])
Y4 = list(datasets["32"].iloc[:, 3])
Y5 = list(datasets["64"].iloc[:, 3])
Y6 = list(datasets["128"].iloc[:, 3])

plt.plot(X, Y0, label="2GB RAM cache")
plt.plot(X, Y1, label="4GB RAM cache")
plt.plot(X, Y2, label="8GB RAM cache")
plt.plot(X, Y3, label="16GB RAM cache")
plt.plot(X, Y4, label="32GB RAM cache")
plt.plot(X, Y5, label="64GB RAM cache")
plt.plot(X, Y6, label="128GB RAM cache")
plt.ylim(0)
plt.title("Cold Starts (Disk 500GB 10GB/s R/W)")
plt.xlabel("Second")
plt.ylabel("Number of cold invocations")
plt.legend()
plt.show()
```

```python
#Plot warm starts vs RAM
data = pd.read_csv("warm.csv")
df = pd.DataFrame(data)
X = list(df.iloc[:, 0])
Y = list(df.iloc[:, 1])
X = [x/1000 for x in X]
plt.plot(X, Y, color="orange")
plt.ylim(0)
plt.title("RAM Cache size vs warm starts (Disk 500GB 10GB/s R/W)")
plt.xlabel("RAM cache size (GB)")
plt.ylabel("Percentage of warm starts")
```

```python
#Plot latency CDFs
ram_2 = pd.read_csv("latencies_ram_2.csv")
ram_4 = pd.read_csv("latencies_ram_4.csv")
ram_8 = pd.read_csv("latencies_ram_8.csv")
ram_16 = pd.read_csv("latencies_ram_16.csv")
ram_32 = pd.read_csv("latencies_ram_32.csv")
ram_64 = pd.read_csv("latencies_ram_64.csv")
ram_128 = pd.read_csv("latencies_ram_128.csv")
df_2 = pd.DataFrame(ram_2)
df_4 = pd.DataFrame(ram_4)
df_8 = pd.DataFrame(ram_8)
df_16 = pd.DataFrame(ram_16)
df_32 = pd.DataFrame(ram_32)
df_64 = pd.DataFrame(ram_64)
df_128 = pd.DataFrame(ram_128)
x2 = np.sort(list(df_2.iloc[:, 0]))
y2 = np.arange(len(df_2)) / float(len(df_2))
x4 = np.sort(list(df_4.iloc[:, 0]))
y4 = np.arange(len(df_4)) / float(len(df_4))
x8 = np.sort(list(df_8.iloc[:, 0]))
y8 = np.arange(len(df_8)) / float(len(df_8))
x16 = np.sort(list(df_16.iloc[:, 0]))
y16 = np.arange(len(df_16)) / float(len(df_16))
x32 = np.sort(list(df_32.iloc[:, 0]))
y32 = np.arange(len(df_32)) / float(len(df_32))
x64 = np.sort(list(df_64.iloc[:, 0]))
y64 = np.arange(len(df_64)) / float(len(df_64))
x128 = np.sort(list(df_128.iloc[:, 0]))
y128 = np.arange(len(df_128)) / float(len(df_128))

# plotting
plt.xlabel('latency (ms)')
plt.ylabel('percentile')
  
plt.title('Latency CDF (Disk 500GB 10GB/s R/W)')
  
plt.plot(x2, y2, label="2GB RAM Cache")
plt.plot(x4, y4, label="4GB RAM Cache")
plt.plot(x8, y8, label="8GB RAM Cache")
plt.plot(x16, y16, label="16GB RAM Cache")
plt.plot(x32, y32, label="32GB RAM Cache")
plt.plot(x64, y64, label="64GB RAM Cache")
plt.plot(x128, y128, label="128GB RAM Cache")
plt.ylim(0)
plt.xlim(0)
plt.legend()
plt.show()
```
